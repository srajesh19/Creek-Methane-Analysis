```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

# Import seaborn
import seaborn as sns
# Apply the default theme
sns.set_theme()

df = pd.read_csv("/content/Combined - Compiled (3).csv")
```

```python
df.columns
```

```python
df = df.replace('-' , np.nan)
df['MAIN PPM'] = df['MAIN PPM'].astype(float)
df['DATE']= pd.to_datetime(df['DATE'])
ax = df.plot.scatter('DATE','MAIN PPM',figsize=(20,10),s=100)
ax.set_xlabel('DATE', fontsize=32)
ax.set_ylabel('MAIN PPM', fontsize=32)
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
	label.set_fontsize(16)
```

```python
df['COMBINED PPM'] = df['COMBINED PPM'].astype(float)
df['RUNOFF PPM'] = df['RUNOFF PPM'].astype(float)

ax1 = df.plot(kind='scatter', x='DATE', y='MAIN PPM', color='r',figsize=(20,10), s=100)
ax2 = df.plot(kind='scatter', x='DATE', y='RUNOFF PPM', color='g', ax=ax1, s=100)
ax3 = df.plot(kind='scatter', x='DATE', y='COMBINED PPM', color='b', ax=ax1, s=100)
ax1.legend(["Main","Runoff","Combined"],prop={"size":24})
ax1.set_xlabel('DATE', fontsize=32)
ax1.set_ylabel('PPM', fontsize=32)
for label in (ax1.get_xticklabels() + ax1.get_yticklabels()):
	label.set_fontsize(16)
```
